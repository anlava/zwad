```python
#from updated from ANOMALIES_PAD 11/23/2020

m31_maria_anomalies_PAD = pd.read_csv('../data/m31_maria.csv')
disk_maria_anomalies_PAD = pd.read_csv('../data/disk_maria.csv')
deep_maria_anomalies_PAD = pd.read_csv('../data/deep_maria.csv', encoding='latin-1') #was throwing weird error

deep_maria_anomalies_PAD.head()
```

```python
# Set ANOMALIES tag; i.e. "uncat"

anomalies_tag_1 = 'uncat'

m31_anomalous_tag = m31_maria_anomalies_PAD[m31_maria_anomalies_PAD.tag == '{}'.format(anomalies_tag_1)]
disk_anomalous_tag = disk_maria_anomalies_PAD[disk_maria_anomalies_PAD.tag == '{}'.format(anomalies_tag_1)]
deep_anomalous_tag = deep_maria_anomalies_PAD[deep_maria_anomalies_PAD.tag == '{}'.format(anomalies_tag_1)]
deep_anomalous_tag
```

```python
#remember to format cells -> number in Excel 

#M31
m31_maria_anomalies_PAD_oid = m31_anomalous_tag['oid']
m31_expert_anomalies = m31.loc[m31_maria_anomalies_PAD_oid]
m31_expert_anomalies.to_csv('../data/m31_expert_anomalies.csv', index_label='oid') 

#Disk
disk_maria_anomalies_PAD_oid = disk_anomalous_tag['oid']
disk_expert_anomalies = disk.loc[disk_maria_anomalies_PAD_oid]
disk_expert_anomalies.to_csv('../data/disk_expert_anomalies.csv', index_label='oid') 

#Deep
deep_maria_anomalies_PAD_oid = deep_anomalous_tag['oid']
deep_expert_anomalies = deep.loc[deep_maria_anomalies_PAD_oid]
deep_expert_anomalies.to_csv('../data/deep_expert_anomalies.csv', index_label='oid') 

m31_expert_anomalies #remember to format cells -> number if you want to see full oid in Excel
```
